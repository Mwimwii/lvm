```python
REF_TEXT="A string is defined as a sequence of characters, and it is important to remember that these are case sensitive. Strings can contain anything from standard alphabet letters, the special symbols, spaces and numerical digits."
REF_PROMPT_WAV="/content/vo/speaker.wav"
CFG=2.7
```

```python
import os
try:
  print(os.listdir())
  os.mkdir('split')
except Exception as e:
  print('WAV splits folder have been created')
finally:
  print('Done')
```

```python
import soundfile as sf
from voxcpm import VoxCPM

model = VoxCPM.from_pretrained("openbmb/VoxCPM2")

wav = model.generate(
    text="How you design a spec depends on the scope of the function. Helper Functions: If a function is private and only used by you within one module, a precondition is fine because you control every single place that function is called. Public/Library Functions: If the function is exported for others to use, you shouldn't rely on preconditions. Instead, you should specify that the function throws an exception for bad inputs, helping the caller identify their own bugs quickly.",
    prompt_wav_path=REF_PROMPT_WAV,      # optional: path to a prompt speech for voice cloning
    prompt_text=REF_TEXT,          # optional: reference text
    cfg_value=CFG,             # LM guidance on LocDiT, higher for better adherence to the prompt, but maybe worse
    inference_timesteps=10,   # LocDiT inference timesteps, higher for better result, lower for fast speed
    normalize=True,           # enable external TN tool
    denoise=True,             # enable external Denoise tool
    retry_badcase=True,        # enable retrying mode for some bad cases (unstoppable)
    retry_badcase_max_times=3,  # maximum retrying times
    retry_badcase_ratio_threshold=6.0, # maximum length restriction for bad case detection (simple but effective), it could be adjusted for slow pace speech
)

# Attempt to create the folder
try:
  os.mkdir('split')
except Exception as e:
  print('WAV splits folder have been created')
finally:


sf.write("output.wav", wav, 48000)
print("saved: output.wav")
```
